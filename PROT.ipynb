```python
def prot_translate(seq):
    seq = seq.lower().replace('\n', '').replace(' ', '')
    #printing sequence
    peptide = ''
    
    for i in range(0, len(seq), 3):
        cod = seq[i: i+3]
        amino_acid = RNA_codon_table.get(cod, '*')
        #printing amino_acid
        if amino_acid != '*':
            peptide += amino_acid
        else:
            break
                
    return peptide

nuc_base = ['u', 'c', 'a', 'g']
cod = [a+b+c for a in nuc_base for b in nuc_base for c in nuc_base]
aa = 'FFLLSSSSYY**CC*WLLLLPPPPHHQQRRRRIIIMTTTTNNKKSSRRVVVVAAAADDEEGGGG'
RNA_codon_table = dict(zip(cod, aa))


f = open('rosalind_prot.txt','r')
s = f.readline().strip()
print (prot_translate(s))
```
