```python
import Bio
from Bio import motifs
from Bio.Seq import Seq
import Bio.Alphabet

one = open("rosalind_cons.txt", 'r')
data = one.readlines()

main = {}
key = ''

for char in data:
    char = char.strip()
    if char[0] == '>':
        key = char[1:]
        main[key] = '' 
    else:
        main[key] += char

instances = []
for seq in main.values():
    instances.append(Seq(seq))
mot = motifs.create(instances)


cons_prof = mot.counts 
consensus = mot.consensus

print (consensus)
for key, value in cons_prof.items():
    print ( "".join(key + ": " + " ".join(map(str, value))))
```
